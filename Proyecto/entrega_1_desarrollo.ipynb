```python
X_t0_path = "competition_files/X_t0.parquet"
y_t0_path = "competition_files/y_t0.parquet"
X_t1_path = "competition_files/X_t1.parquet"
```

```python
x_t0 = pd.read_parquet(X_t0_path)
x_t0
```

```python
for col in x_t0.columns:
    if "timestamp_diff" in col:
        x_t0[col] = pd.to_timedelta(x_t0[col])
    elif "timestamp" in col:
        x_t0[col] = pd.to_datetime(x_t0[col])
    print(f"{col}: {x_t0[col].dtype}")
```
