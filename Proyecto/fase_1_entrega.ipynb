```python
x_t0 = x_t0.drop(columns=['wallet_address'])
```

```python

for col in x_t0.columns:
    if pd.api.types.is_numeric_dtype(x_t0[col]):
        variance = x_t0[col].var()
        print(f"Variance of column {col}: {variance}")
    else:
        print(f"Column {col} is not numeric, cannot calculate variance.")
```

#EliminaciÃ³n de variables con poca varianza

```python

variances = {}
for col in x_t0.columns:
    if pd.api.types.is_numeric_dtype(x_t0[col]):
        variances[col] = x_t0[col].var()

variance_threshold = 0.1

columns_to_drop = [col for col, var in variances.items() if var < variance_threshold]

x_t0_reduced = x_t0.drop(columns=columns_to_drop)

print(f"Columns dropped: {columns_to_drop}")

x_t0_reduced
```

#EliminaciÃ³n de variables altamente correlacionadas

```python

correlation_matrix = x_t0_reduced.corr()

correlation_threshold = 0.8

highly_correlated_vars = []
for i in range(len(correlation_matrix.columns)):
    for j in range(i + 1, len(correlation_matrix.columns)):
        if abs(correlation_matrix.iloc[i, j]) > correlation_threshold:
            highly_correlated_vars.append((correlation_matrix.columns[i], correlation_matrix.columns[j], correlation_matrix.iloc[i,j]))

print("Highly correlated variables:")
for var1, var2, correlation in highly_correlated_vars:
    print(f"{var1} and {var2}: {correlation}")
```

```python

columns_to_drop_corr = set()
for var1, var2, correlation in highly_correlated_vars:
    columns_to_drop_corr.add(var2)

x_t0_reduced = x_t0_reduced.drop(columns=list(columns_to_drop_corr))

print("Shape after dropping highly correlated columns:", x_t0_reduced.shape)
x_t0_reduced
```

#XGBoost

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
import xgboost as xgb


X_train, X_test, y_train, y_test = train_test_split(x_t0_reduced, y_t0, test_size=0.3, random_state=100)

model = xgb.XGBClassifier(objective='binary:logistic', random_state=100)
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

print(classification_report(y_test, y_pred))
```

```python
import optuna
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
import lightgbm as lgb

def objective(trial):
    classifier_name = trial.suggest_categorical('classifier', ['XGBoost', 'LightGBM', 'RandomForest', 'SVC', 'KNN'])
    
    if classifier_name == 'XGBoost':
        param = {
            'objective': 'binary:logistic',
            'eval_metric': 'logloss',
            'booster': 'gbtree',
            'lambda': trial.suggest_loguniform('lambda', 1e-3, 10.0),
            'alpha': trial.suggest_loguniform('alpha', 1e-3, 10.0),
            'subsample': trial.suggest_uniform('subsample', 0.4, 1.0),
            'colsample_bytree': trial.suggest_uniform('colsample_bytree', 0.4, 1.0),
            'learning_rate': trial.suggest_loguniform('learning_rate', 1e-3, 0.1),
            'n_estimators': trial.suggest_int('n_estimators', 100, 1000),
            'max_depth': trial.suggest_int('max_depth', 1, 20),
            'random_state': 100
        }
        model = xgb.XGBClassifier(**param, n_jobs=8)
    
    elif classifier_name == 'LightGBM':
        param = {
            'objective': 'binary',
            'metric': 'binary_logloss',
            'lambda_l1': trial.suggest_loguniform('lambda_l1', 1e-3, 10.0),
            'lambda_l2': trial.suggest_loguniform('lambda_l2', 1e-3, 10.0),
            'num_leaves': trial.suggest_int('num_leaves', 2, 256),
            'feature_fraction': trial.suggest_uniform('feature_fraction', 0.4, 1.0),
            'bagging_fraction': trial.suggest_uniform('bagging_fraction', 0.4, 1.0),
            'learning_rate': trial.suggest_loguniform('learning_rate', 1e-3, 0.1),
            'n_estimators': trial.suggest_int('n_estimators', 100, 1000),
            'max_depth': trial.suggest_int('max_depth', 1, 20),
            'random_state': 100
        }
        model = lgb.LGBMClassifier(**param, n_jobs=8)
    
    elif classifier_name == 'RandomForest':
        param = {
            'n_estimators': trial.suggest_int('n_estimators', 100, 1000),
            'max_depth': trial.suggest_int('max_depth', 1, 20),
            'min_samples_split': trial.suggest_int('min_samples_split', 2, 20),
            'min_samples_leaf': trial.suggest_int('min_samples_leaf', 1, 20),
            'random_state': 100
        }
        model = RandomForestClassifier(**param, n_jobs=8)
    
    elif classifier_name == 'SVC':
        param = {
            'C': trial.suggest_loguniform('C', 1e-3, 10.0),
            'kernel': trial.suggest_categorical('kernel', ['linear', 'poly', 'rbf', 'sigmoid']),
            'gamma': trial.suggest_categorical('gamma', ['scale', 'auto']),
            'random_state': 100
        }
        model = SVC(**param)
    
    elif classifier_name == 'KNN':
        param = {
            'n_neighbors': trial.suggest_int('n_neighbors', 1, 20),
            'weights': trial.suggest_categorical('weights', ['uniform', 'distance']),
            'p': trial.suggest_int('p', 1, 2)
        }
        model = KNeighborsClassifier(**param, n_jobs=8)
    
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    f1 = classification_report(y_test, y_pred, output_dict=True)['1']['f1-score']
    return f1

study = optuna.create_study(direction='maximize')
study.optimize(objective, n_trials=100, n_jobs=8, gc_after_trial=True)

print("Best trial:")
trial = study.best_trial

print("  Value: {}".format(trial.value))
print("  Params: ")
for key, value in trial.params.items():
    print("    {}: {}".format(key, value))
```
