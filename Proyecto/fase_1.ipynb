```python
from sklearn.base import BaseEstimator, TransformerMixin


class IQRTransformer(BaseEstimator, TransformerMixin):
    def __init__(self, columns, iqr_factor=1.5):
        self.columns = columns
        self.iqr_factor = iqr_factor

    def fit(self, X, y=None):
        self.iqr_bounds = {}
        for col in self.columns:
            Q1 = X[col].quantile(0.25)
            Q3 = X[col].quantile(0.75)
            IQR = Q3 - Q1
            self.iqr_bounds[col] = (Q1 - self.iqr_factor * IQR, Q3 + self.iqr_factor * IQR)
        return self

    def transform(self, X, y=None):
        X = X.copy()
        for col, bounds in self.iqr_bounds.items():
            X[col] = np.clip(X[col], bounds[0], bounds[1])
        return X
```

```python
class StdTransformer(BaseEstimator, TransformerMixin):
    def __init__(self, columns, std_threshold=3):
        self.columns = columns
        self.std_threshold = std_threshold

    def fit(self, X, y=None):
        self.std_stats = {}
        for col in self.columns:
            mean = X[col].mean()
            std = X[col].std()
            self.std_stats[col] = (mean - self.std_threshold * std, mean + self.std_threshold * std)
        return self

    def transform(self, X, y=None):
        X = X.copy()
        for col, bounds in self.std_stats.items():
            X[col] = np.clip(X[col], bounds[0], bounds[1])
        return X
```
